```python
#find out location
COL = 'Location'

def not_blank(inp):
    if len(''.join([str(i) for i in inp])) >= 1:
        return True
    else:
        return False

def make_loc(row):
    if not_blank(row[[i for i in range(1,22,2)]]):
        return 'Central'
    elif not_blank(row[[i for i in range(2,23,2)]]): 
        return 'Outside'
    else:
        return 'Blank'

r = pd.DataFrame(list(zip(df['id'], df.apply(lambda row: make_loc(row),axis=1))), columns = ('id', COL))

org = org.merge(r, on = 'id')

org.groupby(COL).size()
```

```python
#find out type working for

# Government of Nepal - Working at central level	Government of Nepal - Working at sub-national, district/VDC level
# International NGO - Working at central level	International NGO - Working at sub-national, district/VDC level	
# National / Local NGO - Working at central level	National / Local NGO - Working at sub-national, district/VDC level
# Donors - Working at central level	Donors - Working at sub-national, district/VDC level	
# Development Partners - Working at central level	Development Partners - Working at sub-national, district/VDC level	
# UN - Working at central level	UN - Working at sub-national, district/VDC level	
# Private Sector - Working at central level	Private Sector - Working at sub-national, district/VDC level	
# Academic or Research Institution - Working at central level	Academic or Research Institution - Working at 
# Beneficiaries’ Federations /Associations - Working at central level	Beneficiaries’ Federations 
# Media - Working at central level	Media - Working at sub-national, district/VDC level	
# Others - Working at central level	Others - Working at sub-national, district/VDC level
COL = 'Work Type'


def make_org(row):
    if row[3] or row[4] != '':
        return 'GoN'
    elif row[5] or row[6] != '':
        return 'INGO'
    elif row[7] or row[8] != '':
        return 'Local NGO'
    elif row[9] or row[10] != '':
        return 'Donor'
    elif row[11] or row[12] != '':
        return 'Dev Partner'
    elif row[13] or row[14] != '':
        return 'UN'
    elif row[16] or row[16] != '':
        return 'Private'
    elif '' not in row[17:24].values:
        return 'Other'
    
    
r = pd.DataFrame(list(zip(df['id'], df.apply(lambda row: make_org(row),axis=1))), columns = ('id', COL))

org = org.merge(r, on = 'id')

org.groupby(COL).size()
```

```python
#most valued - set all values to 1
ed = df.filter(regex='^id|un_value')

def setone(x):
    if x != '' and len(str(x)) < 3:
        return 1
    else:
        return x

ed = ed.applymap(setone)

org = org.merge(ed, on = 'id')
```

```python
#top themes
import re

ed = df.filter(regex='^id|un_prior(.*) - 1')

#get the names of values to use and remove leading/trailing junk
gc = []
for i in ed.columns[:-1]:
    if i.startswith('un_prior Improving'):
        gc.append(('un_prior Improving access to quality basic social services',''))
    else:
        r = re.search('(.*?) -..$', str(i))
        if r:
            gc.append(r.groups(0))
        else:
            pass

#do some silly transformations
gc = [x[0] for x in list(set(gc))]

mrg = ed.filter(regex='^id')

#add in columns to main dataset that we will be adding to
for i in gc:
    mrg[i] = None

#iterate through each column whose name matches lookup and see if a response was given 
#...for just the highest and apply if so

for row in ed.iterrows():
    for v in gc:
        vals = row[1].filter(regex=v).tolist()
        if len(''.join([str(i) for i in vals])) > 0:
            mrg.loc[mrg.id == row[1]['id'],v] = 1
        else:
            mrg.loc[mrg.id == row[1]['id'],v] = None

def sum_wrk(df):
    for col in df.columns.tolist():
        print(np.sum(df[col]))

a = mrg.sum(axis=0)[1:]

a

org = org.merge(mrg, on = 'id')

#see values descending
# a.sort_values(ascending=False)

# #plot sums
# a = mrg.sum(axis=0)[1:]

# objects = a.index.tolist()
# y_pos = np.arange(len(a))
# performance = a
 
# plt.bar(y_pos, performance, width=.8)
# plt.xticks(y_pos, objects, rotation=90)

# plt.show()
```

```python
#UN ranking - get unique value for each question group.
import re

ed = df.filter(regex='^id|un_position')
ed.columns = ed.columns.str.replace('\xa0', '')
ed.columns = ed.columns.str.replace('\xa0', '')
ed.columns = ed.columns.str.replace('(', '')
ed.columns = ed.columns.str.replace(')', '')
ed.columns = ed.columns.str.replace('[', '')
ed.columns = ed.columns.str.replace(']', '')


#get the names of values to use and remove leading/trailing junk
gc = []
for i in ed.columns[1:-1]:
    r = re.search('(.*?) -.*$', str(i))
    if r:
        gc.append(r.groups(0))
    else:
        gc.append('missing' + i) 

gc = [x[0] for x in list(set(gc))]
# gc.remove('m')
# gc.remove('Others')

mrg = ed.filter(regex='^id')

#add in columns to main dataset that we will be adding to
for i in gc:
    mrg[i] = ''

#order: VWell, W, MW, NW, Not at all, DK
#iterate through each column whose name matches lookup and see if a response was given for any of the 3 and apply if so
for row in ed.iterrows():
    for v in gc:
        vals = row[1].filter(regex=v).tolist()
        r = ''.join([str(i) for i in vals])
        if len(r) > 0:
            mrg.loc[mrg.id == row[1]['id'],v] = int(r)
        else:
            mrg.loc[mrg.id == row[1]['id'],v] = ''

org = org.merge(mrg, on = 'id')
```

```python
#areas of improvement (only count of strongly agree/disagree)

ed = df.filter(regex='^id|un_agree.*(Agree|Strongly agree)')
ed.columns = ed.columns.str.replace('(', '')
ed.columns = ed.columns.str.replace(')', '')

#get the names of values to use and remove leading/trailing junk
gc = []
for i in ed.columns[1:-1]:
    r = re.search('(.*?) -.*$', str(i))
    if r:
        gc.append(r.groups(0))
    else:
        gc.append('missing' + i) 

gc = [x[0] for x in list(set(gc))]

mrg = ed.filter(regex='^id')

#add in columns to main dataset that we will be adding to
for i in gc:
    mrg[i] = ''

#iterate through each column whose name matches lookup and see if a response was given for either and apply 1 if so
for row in ed.iterrows():
    for v in gc:
        vals = row[1].filter(regex=v).tolist()
        r = ''.join([str(i) for i in vals])
        if len(r) > 0:
            mrg.loc[mrg.id == row[1]['id'],v] = 1
        else:
            mrg.loc[mrg.id == row[1]['id'],v] = ''
            
org = org.merge(mrg, on = 'id')
```

```python
#set all blanks to 0, remove unncessary columns

org = org.applymap(lambda x: 0 if x == '' else x)
```

```python
#most valued services
KEY = '^id|^Work Type|^Location|un_value'

an = org.filter(regex=KEY)
an = an[an.columns[:-1]]

#delete non respondants

c=0
for row in an.iterrows():
    if sum(row[1].values[3:]) == 0:
        c+=1
        an = an.drop(an[an['id'] == row[1].values[0]].index)
        r = row

# top 5 diff by location
loc_diff = pd.DataFrame(an.groupby(['Location']).mean().T)[1:]
loc_diff['Difference'] = abs(loc_diff['Central'] - loc_diff['Outside'])
loc_diff = loc_diff.merge(an.mean()[1:].T.to_frame(), left_index=True, right_index=True).rename(columns = {0 : 'Total'})

loc_diff.sort_values(by = 'Difference', ascending=False)[:5]

# top 5 most polarizing questions [highest SD]
type_diff = pd.DataFrame(an.groupby(['Work Type']).mean().T)[1:]
type_diff['Average Difference'] = np.std(type_diff[type_diff.columns], axis = 1)
type_diff = type_diff.merge(an.mean()[1:].T.
                            to_frame(), left_index=True, right_index=True).rename(columns = {0 : 'Total'})

#show
loc_diff.sort_values(by = 'Total', ascending=False)
```

```python
# for final report

l = an.groupby(['Location']).sum().ix[:, 1:].T
l['cent_pct'] = l['Central']/(l['Central'] + l['Outside'])
l['out_pct'] = l['Outside']/(l['Central'] + l['Outside'])


g = an.groupby(['Work Type']).sum().ix[:, 1:].T

for v in (g.columns):
    g[v + '_pct'] = g[v] / sum([g[i] for i in g.columns])

g['sum'] = sum([g[i] for i in g.columns if '_pct' not in i])

g.sort_values(by = 'sum', ascending=False).to_clipboard(excel=True, sep=';')

l
```
