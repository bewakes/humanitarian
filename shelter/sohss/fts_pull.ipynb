```python
"""read in SC historical data, convert ISOz"""
LOC = '/Users/ewanog/Google Drive/SoHSS/Report/Docs/sc_historical.xlsx'
hist = pd.read_excel(LOC, sheet_name = 'Sheet2')

#fix columns
hist.rename(columns = {'Country ' : 'Country'}, inplace = True)

#Pacific Region given for 2015 - 2018, remove
hist = hist[hist['Country'] != 'Pacific Region']

#Rename 'Chili' to 'Chile'
hist.loc[hist['Country'] == 'Chili', 'Country'] = 'Chile'

#add ISO codes
hist['ISO3'] = hist['Country'].apply(lambda x : coco.convert(names = x, to = 'ISO3'))

#total number of crisis: 230
print(len(hist))
```

```python
"""group by year, country and pull down total values with optional filter by cluster"""
countries = json_normalize(json.loads(urllib.request.urlopen("https://api.hpc.tools/v1/public/location")
                                      .read().decode()))['data']

print('pulled countries')

hist['fts_url'] = hist.apply(lambda x : 'https://api.hpc.tools/v1/public/fts/flow?countryISO3={0}&year={1}&globalClusterCode=shl'
                             .format(x['ISO3'], x['Year']), axis = 1)

hrefz = hist['fts_url']

def exception_handler(request, exception):
     print('Bad URL for ' + request)

resps = []
rs = (grequests.get(ref) for ref in hrefz)
resps += grequests.map(rs, exception_handler = exception_handler, size = 25)

good_resps = []
bad_resps = []
for r in resps:
    if r.status_code == 200:
        good_resps.append(json.loads(r.content))
    else:
        bad_resps.append(json.loads(r.content))


j = json_normalize(good_resps).add_prefix('fts.')
```
