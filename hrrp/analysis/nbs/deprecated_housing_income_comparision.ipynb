```python
"""helper methods for district code, 12 digit, correctly formatting vdc label"""

codes = pcode['VDC_CODE']
POSS = {'5': [v[1]['VDC_NAME'] for v in pcode.iterrows() if v[1]['VDC_CODE'][12] == '5'],
        '3': [v[1]['VDC_NAME'] for v in pcode.iterrows() if v[1]['VDC_CODE'][12] == '3'],
        '9': [v[1]['VDC_NAME'] for v in pcode.iterrows() if v[1]['VDC_CODE'][12] == '9'],
        '2': [v[1]['VDC_NAME'] for v in pcode.iterrows() if v[1]['VDC_CODE'][12] == '2'],
        '1': [v[1]['VDC_NAME'] for v in pcode.iterrows() if v[1]['VDC_CODE'][12] == '1']
       }

def get_12th_match(name):
    """get in name, return reccomended 12th digit in (name , result)"""
    recs = []
    for k,v in POSS.items():
        if name in v:
            recs.append(k)
            
    if len(set(recs)) > 1:
        return 'too many matches for ' + name
    elif len(set(recs)) == 0:
        return 'no matches for ' + name
    else:
        return str(recs[0])

def get_12th_approx(name):
    if 'kathmandu metropolitan' in name.lower():
        return '1'
    elif len([True for v in ['sub metropolitan', 'sub-metropolitan'] if v in name.lower()]) >= 1:
        return '2'
    elif 'municipality' in name.lower():
        return '3'
    elif len([True for v in ['park', 'wildlife', 'forest', 'development'] if v in name.lower()]) >= 1:
        return '9'
    else:
        return '5'
    
#test    
#TODO: make actual test
l = []
    
for v in pcode['VDC_NAME']:
    l.append(get_12th_approx(v))
    
test_cnts = pd.Series(l).value_counts()
act_cnts = pd.Series([v[12] for v in codes]).value_counts()

for k, v in act_cnts.items():
    if test_cnts[k] != v:
        print('not match!')
        print(test_cnts)
        print(act_cnts)
        break
    
# pd.Series([v[12] for v in codes]).value_counts()
# """
# 5 = normal
# 3 = Metropolitan
# 9 = Wildlife things
# 2 = like 'Sub Metropolitan'
# 1 = like 'Kathmandu Metropolitan'
# """


#districts
def get_dist(dist_in):
    """get district code"""
    try:
        return dist_codes[dist_codes['DIST_NAME'] == dist_in]['DIST_CODE'].values[0]
    except:
        return 'NF'
#     except Exception as e:
#         raise ValueError('Name {0} not found'.format(dist_in))


#format vdc num (length of 3)
def get_vdc_digit(vdc_num):
    """ie 5 > 005, 40 > 040"""
    vdc_num = str(vdc_num)
    return '0' * (3 - len(vdc_num)) + vdc_num
```
